```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)


# Read in the data

df = pd.read_csv(r"D:\Web Scraping\Python project\movies.csv")
df
```

```python
df.head()
```

```python
# lets see if there is any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# datatype for our columns

df.dtypes
```

```python
# filling 0 on nan values

df['budget'] = df['budget'].fillna(0)
df['gross'] = df['gross'].fillna(0)
```

```python
# changing datatype of budget column and 'astype' changes the type

df['budget'] =df['budget'].astype('int64')

df['gross'] =df['gross'].astype('int64')
```

```python
df.head()
```

```python
# create correct year column

df['yearcorrect'] = df['released'].astype(str).str[:4]
df.head()
```

```python
df = df.sort_values (by = ['gross'], inplace = False, ascending = False )
```

```python
#we can see entire code with 
# pd.set_option('display.max_rows',None)                                
```

```python
# drop any duplicates
df['company'] = df['company'].drop_duplicates()
```

```python
# scatter plot with budget vs gross revenue
plt.scatter(x = df['budget'], y = df['gross'])

plt.title('Budget vs Gross Earning')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget For Film')
plt.show()
```

```python
# changes everything which was in string into numeric
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'): 
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
df.head()
```

```python
correlation_matrix = df_numerized.corr(numeric_only=True, method ='pearson')
sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
pd.set_option('display.max_rows', None)  # This line displays all
print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```
